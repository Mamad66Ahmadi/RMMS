```python
import sqlite3
```

```python
conn = sqlite3.connect("daily_jobs.db")  # replace with your DB filename
cursor = conn.cursor()
```

```python
# Create 'objects' table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS objects (
        tag TEXT PRIMARY KEY,
        father_tag TEXT,
        unit TEXT,
        special_comment TEXT,
        integral_part TEXT
    )
""")

conn.commit()
#conn.close()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables:", tables)
```

```python
cursor.execute("PRAGMA table_info(objects);")
schema = cursor.fetchall()
print("Schema:", schema)
```

```python
try:
    cursor.execute("ALTER TABLE objects ADD COLUMN modified_by TEXT;")
except sqlite3.OperationalError as e:
    print("created_by already exists:", e)
```

```python
cursor.execute("SELECT * FROM objects;")
rows = cursor.fetchall()

for row in rows:
    print(row)
```
