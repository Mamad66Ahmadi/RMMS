# Routes

```python
excel_path = "01 Routes after removing2.xlsx"   # üîπ your Excel file
routes_df = pd.read_excel(excel_path)
routes_df
```

```python
# --- 6Ô∏è‚É£ Drop old 'routes' table if exists ---
cursor.execute("DROP TABLE IF EXISTS routes")

# --- 7Ô∏è‚É£ Create 'routes' table ---
cursor.execute("""
CREATE TABLE routes (
    Route_ID INTEGER PRIMARY KEY,
    PMRoute_Code TEXT,
    PMRoute_Desc TEXT,
    Object_Tag TEXT,
    StandardJob_Desc TEXT
)
""")

# --- 8Ô∏è‚É£ Insert data into 'routes' table ---
routes_df.to_sql("routes", conn, if_exists="append", index=False)

# --- 9Ô∏è‚É£ Confirm import ---
count = cursor.execute("SELECT COUNT(*) FROM routes").fetchone()[0]
print(f"‚úÖ Successfully imported {count} rows into 'routes' table.")
```
