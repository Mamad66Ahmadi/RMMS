```python
import sqlite3
import hashlib
import os

DB_PATH = "data/daily_jobs.db"

def hash_password(password):
    salt = os.urandom(16).hex()
    hashed = hashlib.sha256((salt + password).encode()).hexdigest()
    return f"{salt}:{hashed}"

username_to_reset = "ahmdi-sm"
new_password = "1234"  # whatever you want

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()
cur.execute("UPDATE users SET password=? WHERE username=?", (hash_password(new_password), username_to_reset))
conn.commit()
conn.close()

print(f"Password for {username_to_reset} reset successfully!")
```
