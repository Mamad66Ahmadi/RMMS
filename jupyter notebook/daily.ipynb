```python
import streamlit as st
from datetime import datetime

st.set_page_config(page_title="Daily Jobs", layout="wide")

def qp_first(key: str, default: str = "Unknown") -> str:
    """Return the first query-param value for `key`.
    Handles both list-return and string-return behavior of st.query_params.
    """
    val = st.query_params.get(key, default)
    # If streamlit returns a list (expected), pick first element
    if isinstance(val, list):
        return val[0] if val else default
    # If it returned a string unexpectedly, return it (not its first char)
    return val or default

st.title("ğŸ“‹ Daily Jobs Page")

# --- Date ---
today = datetime.today().strftime("%A, %B %d, %Y")
st.markdown(
    f"<div style='text-align:right; color:#555;'>{today}</div>",
    unsafe_allow_html=True
)

# --- Get query parameters safely ---
username = qp_first("username", "Unknown")
name = qp_first("name", "")
department = qp_first("department", "")

if username == "Unknown" or not username.strip():
    st.warning("âš ï¸ You should log in first.")

    # Link back to Log in page
    st.markdown(
        "<a href='/' style='font-size:16px; color:blue; text-decoration:underline;'>â¬…ï¸ Go to Log-in Page</a>",
        unsafe_allow_html=True
    )
else:
    # --- Show user info (supports Persian + English) ---
    st.success(f"âœ… Logged in as: {username}")

    # Use RTL styling for name if it likely contains Persian â€” the container will still display English fine.
    st.markdown(
        f"""
        <div style="margin-top:15px; padding:10px; border:1px solid #ccc; border-radius:10px;">
            <p>ğŸ‘¤ <b>Username:</b> {username}</p>
            <p style="direction:rtl; text-align:left;">ğŸ“› <b>Name:</b> {name}</p>
            <p>ğŸ¢ <b>Department:</b> {department}</p>
        </div>
        """,
        unsafe_allow_html=True
    )

    st.info("Here you can see and manage your daily jobs.")
```
